```python
import spacy
nlp = spacy.load('en_core_web_lg')
```

```python
doc = nlp(u"The cheetah swiftly accelerated towards its prey across the bushes.")
```

```python
len(doc.vocab)
```

```python
from spacy import displacy
```

```python
displacy.render(doc,style='dep',jupyter=True,options={'distance':100,'compact':'True','color':'crimson','bg':'aqua','font':'Helvetica'})
```

```python
doc2 = nlp(u"There are several valuable equipments available in the locker room. They have been kept specifically for students as part of their study material.")
```

```python
spans = list(doc2.sents)
spans
```

```python
doc2.sentiment
```

```python
doc2.similarity(doc)
```

```python
displacy.serve(spans,style='dep',options={'distance':100},port=8000,host='127.0.0.1')
```
