```python
doc8 = nlp("We haven't visited any popular tourism destination so far this year. However, we'll certainly attend Ben's birthday party tomorrow!")
```

```python
for token in doc8:
    print(token)
```

```python
for entity in doc8.ents:
    print(entity)
#     print(entity.lemma_)
    print(entity.label_)
    print('\n')
```

```python
doc7 = nlp("Car insurance companies are offering greater benefits to their customers.")
```

```python
for chunk in doc7.noun_chunks:
    print(chunk)
```

```python
doc10 = nlp("Singapore is really expensive and an average resident over there has a net annual expenditure of around $1568.45 million. Microsoft has a gross revenue of about $78 trillion per annum.")
```

```python
for entity in doc10.ents:
    print(entity)
    print(entity.label_)
    print(str(spacy.explain(entity.label_)))
    print('\n')
```

```python
for chunk in doc10.noun_chunks:
    print(chunk)
    print(chunk.label_)
    print(str(spacy.explain(chunk.label_)))
    print('\n')
```

```python
from spacy import displacy
```

```python
doc11 = nlp("Apple has generated a profit of nearly $8 million over the past quarter of an hour or so as a result of its burgeoning sales of iPods due to its ever-rising demand among customers.")
```

```python
displacy.render(doc11,style='dep',jupyter=True,options={'distance': 80, 'color': 'blue'})
```

```python
doc12 = nlp("Long Short Term Memory(LSTM) networks hold a wide range of applications in the domain of natural language processing(NLP) due to their exceptional ability to remember long term dependencies. It has a boundless scope and everlasting potential to completely enhance the field of artificial intelligence.")
```

```python
displacy.serve(doc12,style='ent',options={'distance': 60},host="127.1.1.0",port=8000)
```

```python
doc13 = nlp("Adam Smith went to the Walmart store to buy a bunch of chocolates for his sister Kathy. He had a pocket money of $50 which he decided to invest for this purpose.")
```

```python
for token in doc13:
    print(token.text, token.pos_, token.dep_, token.lemma_)
```

```python
displacy.render(doc13,style='ent',jupyter=True,options={'distance': 100, 'color': 'indigo'},minify=True)
```

```python
displacy.serve(doc13,style='dep',options={'distance': 115, 'color': 'crimson'},host='127.1.1.0',port=8000)
```

```python
displacy.serve(doc13,style='ent',options={'distance': 95, 'color': 'red'},host='127.1.1.0',port=8000)
```
